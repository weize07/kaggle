```python
import pandas as pd
```

```python
df = pd.DataFrame.from_csv('./data/train.csv')
di = {'A': 0, 'C': 1, 'FV': 2, 'I': 3, 'RH': 4, 'RL': 5, 'RP': 6, 'RM': 7}
di_lot_shape = {'Reg': 0, 'IR1': 1, 'IR2': 2, 'IR3': 3}
di_neighborhood = {'Blmngtn':0,'Blueste':1,'BrDale':2,'BrkSide':3,'ClearCr':4,'CollgCr':5,'Crawfor':6,'Edwards':7,'Gilbert':8,'IDOTRR':9,'MeadowV':10,'Mitchel':11,'Names':12,'NoRidge':13,'NPkVill':14,'NridgHt':15,'NWAmes':16,'OldTown':17,'SWISU':18,'Sawyer':19,'SawyerW':20,'Somerst':21,'StoneBr':22,'Timber':23,'Veenker':24}

df = df[df.MSZoning.isin(di)]
df = df[df.LotShape.isin(di_lot_shape)]
df = df[df.Neighborhood.isin(di_neighborhood)]
# df = df.MSZoning.isin({'MSZoning': di, 'LotShape': di_lot_shape, 'Neighborhood': di_neighborhood})

df = df.replace({"MSZoning": di})
df = df.replace({"LotShape": di_lot_shape})
df = df.replace({"Neighborhood": di_neighborhood})
df
```

```python
%matplotlib inline
df['SalePrice'].hist()
```

```python
# features:
# MSSubClass
# MSZoning
# LotFrontage
# LotArea
# Neighborhood
# OverallQual
# OverallCond
# YearBuilt
# 'YearRemodAdd'
x_df = df[['MSSubClass', 'MSZoning', 'LotFrontage', 'LotArea', 'Neighborhood', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd']]
# x_df = df[['MSZoning']]
y_df = df['SalePrice']
x_df = x_df.fillna(value = 0)

# x_df.apply(pd.to_numeric)
# # features
X = x_df.as_matrix()
Y = y_df.as_matrix()
X.astype(float)
import numpy as np
np.savetxt('filtered.csv', X)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(X, Y) #è®­ç»ƒæ¨¡åž‹
```

```python
dft = pd.DataFrame.from_csv('./data/test.csv')
ids = dft.index.values
di = {'A': 0, 'C': 1, 'FV': 2, 'I': 3, 'RH': 4, 'RL': 5, 'RP': 6, 'RM': 7, 'C (all)': 4}
di_lot_shape = {'Reg': 0, 'IR1': 1, 'IR2': 2, 'IR3': 3}
di_neighborhood = {'Blmngtn':0,'Blueste':1,'BrDale':2,'BrkSide':3,'ClearCr':4,'CollgCr':5,'Crawfor':6,'Edwards':7,'Gilbert':8,'IDOTRR':9,'MeadowV':10,'Mitchel':11,'Names':12,'NoRidge':13,'NPkVill':14,'NridgHt':15,'NWAmes':16,'OldTown':17,'SWISU':18,'Sawyer':19,'SawyerW':20,'Somerst':21,'StoneBr':22,'Timber':23,'Veenker':24, 'NAmes' : 0}


dft = dft.replace({"MSZoning": di})
dft = dft.replace({"LotShape": di_lot_shape})
dft = dft.replace({"Neighborhood": di_neighborhood})
dft = dft[['MSSubClass', 'MSZoning', 'LotFrontage', 'LotArea', 'Neighborhood', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd']]
dft = dft.fillna(value = 0)

result = reg.predict(dft)
```

```python
with open('sub.csv', 'w') as output:
    for i in range(len(ids)):
        output.write(str(ids[i]) + ',' + str(result[i]) + '\n')
```
